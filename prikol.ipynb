```python
import pandas as pd
```

```python
world_cups = pd.read_csv('./world_cups.csv')
```

```python
world_cups.info()
```

```python
world_cups_cleaned = world_cups.dropna()
world_cups
```

```python
world_cups_cleaned = world_cups_cleaned.drop_duplicates()
world_cups_cleaned
```

```python
world_cups_cleaned.head()
```

```python
world_cups_cleaned.tail()
```

```python
sliced_data_loc = world_cups.loc[[1930, 1934], ['Year', 'Host Country', 'Winner']]
sliced_data_iloc = world_cups.iloc[:3, :3]
sliced_data_columns = world_cups[['Year', 'Host Country', 'Winner']]

sliced_data_loc
sliced_data_iloc 
sliced_data_columns
```

```python
years_80s = world_cups_cleaned[(world_cups_cleaned['Year'] >= 1980) & (world_cups_cleaned['Year'] < 1990)]
 years_80s
```

```python
high_goals = world_cups_cleaned[world_cups_cleaned['Goals Scored'] > 150]
high_goals
```

```python
sorted_world_cups = world_cups_cleaned.sort_values(by='Goals Scored', ascending=False)
sorted_world_cups[['Year', 'Goals Scored']]
```

```python
world_cups_cleaned.describe()
```

```python
grouped_winners = world_cups_cleaned.groupby('Winner').size().reset_index(name='Count')
grouped_winners
```

```python
top_5_winners = grouped_winners.sort_values(by='Count', ascending=False).head(5)
top_5_winners
```
